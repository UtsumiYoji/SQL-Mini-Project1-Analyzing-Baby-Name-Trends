# Mini-Project 1: Analyzing Baby Name Trends

```python
import psycopg2

passward = input("Enter passward: ")

dsn = "dbname=sql_mini_project1 host=localhost user=postgres password=" + passward
conn = psycopg2.connect(dsn)
cur = conn.cursor()
```

## Task1

```python
cur.execute(
    "SELECT first_name, SUM(num) "
    "FROM usa_baby_names "
    "GROUP BY first_name "
    "ORDER BY SUM(num) DESC"
    )

cur.fetchall()
```

## Task2. Timeless or trendy?

```python
cur.execute(
    "SELECT first_name, COUNT(*),"
    "CASE "
        "WHEN COUNT(*) > 80 THEN 'Classic' "
        "WHEN COUNT(*) > 50 THEN 'Semi-classic' "
        "WHEN COUNT(*) > 20 THEN 'Semi-trendy' "
        "ELSE 'Trendy' "
    "END AS popularity_type "
    "FROM usa_baby_names "
    "GROUP BY first_name "
    "ORDER BY first_name"
    )

cur.fetchall()
```

## Task 3. Top-ranked female names since 1920Â¶

```python
cur.execute(
    "SELECT "
    "RANK () OVER (ORDER BY SUM(num) DESC) AS name_rank, "
    "first_name, SUM(num) AS name_count "
    "FROM usa_baby_names "
    "WHERE sex = 'F' "
    "GROUP BY first_name "
    "ORDER BY name_rank "
    "LIMIT 10"
)

cur.fetchall()
```

## Task 4. Picking a baby name

```python
cur.execute(
    "SELECT first_name, SUM(num) "
    "FROM usa_baby_names "
    "WHERE sex = 'F' AND year >= 2015 "
    "GROUP BY first_name "
    "ORDER BY SUM(num) DESC "
)

cur.fetchall()
```

## Task 5. The Olivia expansion

```python
cur.execute(
    "SELECT year, first_name, num, "
    "SUM(num) OVER(ORDER BY year) AS cumulative_olivias "
    "FROM usa_baby_names "
    "WHERE first_name = 'Olivia' "
    "ORDER BY year"
)

cur.fetchall()
```

## Task 6. Many males with the same name

```python
cur.execute(
    "SELECT year, MAX(num) as max_num "
    "FROM usa_baby_names "
    "WHERE sex = 'M' "
    "GROUP BY year "
    "ORDER BY year"
)

cur.fetchall()
```

## Task 7. Top male names over the years

```python
cur.execute(
    "SELECT usa_baby_names.year, first_name, num "
    "FROM usa_baby_names "
    "JOIN ("
        "SELECT year, MAX(num) as max_num "
        "FROM usa_baby_names "
        "WHERE sex = 'M' "
        "GROUP BY year "
        "ORDER BY year"
    ") ON num = max_num "
    "ORDER BY usa_baby_names.year DESC"
)

cur.fetchall()
```

## Task 8. The most years at number one

```python
cur.execute(
    "SELECT first_name, COUNT(*) "
    "FROM ("
        "SELECT usa_baby_names.year, first_name, num "
        "FROM usa_baby_names "
        "JOIN ("
            "SELECT year, MAX(num) as max_num "
            "FROM usa_baby_names "
            "WHERE sex = 'M' "
            "GROUP BY year "
            "ORDER BY year"
        ") ON num = max_num "
        "ORDER BY usa_baby_names.year DESC"
    ") "
    "GROUP BY first_name "
    "ORDER BY COUNT(*) DESC "
)

cur.fetchall()
```
